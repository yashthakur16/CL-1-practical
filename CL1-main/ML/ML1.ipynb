```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("wine.csv")
```

```python
X=df.drop(columns=["Wine"])
y=df["Wine"]
```

```python
df
```

```python
plt.figure(figsize=(12,7))
plt.scatter(X[y==1]["Alcohol"],X[y==1]["Malic.acid"],label="Customer Segmentation 1",color="yellow")
plt.scatter(X[y==2]["Alcohol"],X[y==2]["Malic.acid"],label="Customer Segmentation 2",color="red")
plt.scatter(X[y==3]["Alcohol"],X[y==3]["Malic.acid"],label="Customer Segmentation 3",color="green")
plt.xlabel("Alcohol")
plt.ylabel("Malic.acid")
plt.legend()

```

```python
sc=StandardScaler()
X_scaled=sc.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
pca_df=pd.DataFrame(data=X_pca, columns=["PC1","PC2"])
pca_df["Customer Segment"]=y
                    

```

```python
pca_df
```

```python
plt.figure(figsize=(12,7))
plt.scatter(pca_df[y==1]["PC1"],pca_df[y==1]["PC2"],label="Customer Segmentation 1",color="yellow")
plt.scatter(pca_df[y==2]["PC1"],pca_df[y==2]["PC2"],label="Customer Segmentation 2",color="red")
plt.scatter(pca_df[y==3]["PC1"],pca_df[y==3]["PC2"],label="Customer Segmentation 3",color="green")
plt.legend()
```
